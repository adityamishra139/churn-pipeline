```python
# Step 2: Load and inspect CSV
import pandas as pd

df = pd.read_csv("rental_data.csv")
df.head()
```

```python
# Step 3: Clean and transform data

# Replace '-' with NaT (Not a Time)
df['End_Date'] = df['End_Date'].replace('-', pd.NaT)

# Convert date columns to datetime
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
df['End_Date'] = pd.to_datetime(df['End_Date'])

# Handle missing rent by filling with average rent
df['Monthly_Rent'] = pd.to_numeric(df['Monthly_Rent'], errors='coerce')
avg_rent = df['Monthly_Rent'].mean()
df['Monthly_Rent'].fillna(avg_rent, inplace=True)

df.head()
```

```python
from datetime import datetime

# Fill missing End_Date with today (for ongoing rentals)
df['End_Date'].fillna(pd.Timestamp(datetime.now()), inplace=True)

# Calculate Tenure in Days & Months
df['Tenure_Days'] = (df['End_Date'] - df['Start_Date']).dt.days
df['Tenure_Months'] = df['Tenure_Days'] / 30

# Calculate Revenue
df['Revenue'] = df['Monthly_Rent'] * df['Tenure_Months']

df[['Rental_ID', 'Product_Name', 'Tenure_Months', 'Revenue']].head()
```

```python
import sqlite3

# Connect to SQLite DB (creates file if doesn't exist)
conn = sqlite3.connect("rental_data.db")

# Load DataFrame into SQLite table
df.to_sql("rentals", conn, if_exists="replace", index=False)

print("Data loaded into SQLite successfully!")
```

```python
# Example: Total revenue per product
query1 = """
SELECT Product_Name, ROUND(SUM(Revenue), 2) as Total_Revenue
FROM rentals
GROUP BY Product_Name
ORDER BY Total_Revenue DESC
"""

pd.read_sql(query1, conn)
```

```python
# Top 3 longest tenures
query2 = """
SELECT Rental_ID, Product_Name, Tenure_Days
FROM rentals
ORDER BY Tenure_Days DESC
LIMIT 3
"""

pd.read_sql(query2, conn)
```

```python
query3 = """
SELECT
  CASE
    WHEN Tenure_Days < 30 THEN '0-1 Month'
    WHEN Tenure_Days < 60 THEN '1-2 Months'
    WHEN Tenure_Days < 90 THEN '2-3 Months'
    ELSE '3+ Months'
  END AS Tenure_Bucket,
  COUNT(*) AS Rental_Count
FROM rentals
GROUP BY Tenure_Bucket
ORDER BY Rental_Count DESC
"""

result3 = pd.read_sql(query3, conn)
result3
```

```python
result3.plot(kind='bar', x='Tenure_Bucket', y='Rental_Count', title='Rental Duration Buckets', legend=False)
```

```python
query_churn = """
SELECT
  CASE
    WHEN DATE(End_Date) < DATE('now') THEN 'Churned'
    ELSE 'Active'
  END AS Customer_Status,
  COUNT(*) AS Count
FROM rentals
GROUP BY Customer_Status
"""
churn_df = pd.read_sql(query_churn, conn)
churn_df
```

```python
churn_df.plot(kind='bar', x='Customer_Status', y='Count', title='Churn vs Active Rentals', legend=False)
```
